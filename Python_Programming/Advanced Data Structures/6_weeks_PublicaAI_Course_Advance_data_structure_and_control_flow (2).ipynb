### String Manipulation in Python






Strings are one of the most commonly used data types in Python. Python provides powerful tools to manipulate strings efficiently. In this section, we will start by exploring some fundamental string operations.

- String Concatenation (Joining Strings)

Concatenation means joining two or more strings together using the `+` operator.

````python
first_name = "Publica"
last_name = "AI"
full_name = first_name + last_name  # Adding a space between the names
print(full_name)
````

````python
first_name = "Publica"
last_name = "AI"
full_name = first_name + " " + last_name  # Adding a space between the names
print(full_name)
````

-  Slicing Strings

Slicing allows you to extract a portion of a string using indexes.

Before we go fully into slicing a string, let us briefly look at indexing.

In Python, indexing is a way to access specific characters in a string. Each character in a string has an index (position), which allows you to extract substrings efficiently. Python supports both positive indexing (from the start) and negative indexing (from the end), making string manipulation flexible.

 Types of Indexing in Python Strings

1 Positive Indexing (Start to End)

The first character of the string has an index of 0, the second has an index of 1, and so on.
This index increases from left to right.

image.png

````python
# Example: Accessing the first and 5th item in the string

text = "HelloWorld"
print(text[0])  # Accessing 'H'
print(text[4])  # Accessing 'o'
````

2 Negative Indexing (End to Start)

The last character of the string has an index of -1, the second last has an index of -2, and so on.
This index decreases from right to left.

````python
#Example:

text = "HelloWorld"
print(text[-1])  # Accessing 'd'
print(text[-5])  # Accessing 'W'
````

 How String Slicing Works

Slicing allows you to extract a part of a string using the syntax:

`string[start:end:step]`

Where:

- start: The index to start slicing (inclusive).
- end: The index to stop slicing (exclusive).
- step: The number of steps to move (default is 1).

````python
# Example
text = "HelloWorld"
print(text[0:5])   # Extracts 'Hello' (from index 0 to 4)
````

````python
print(text[:5])    # Same as above, as start defaults to 0
````

````python
print(text[5:])    # Extracts 'World' (from index 5 to end)
````

````python
print(text[::2])   # Skips every second letter
````

````python
text = "Data Science"
print(text[4])
````

````python
print(text[:4])

````

````python
print(text[5:])   # Extracts 'Science' (from index 5 to end)
````

````python
print(text[-7:-1]) # Extracts 'Scienc' (negative indexing)
````

### String Formatting in Python

String formatting in Python allows you to dynamically insert values into strings in a clean and readable way. Python provides multiple methods for formatting strings, including:

- `f-strings` (formatted string literals)  Recommended for modern Python.

- `.format()` method  Versatile and widely supported.

#### Using f-strings (Formatted String Literals)

`f-strings` (introduced in Python 3.6) provide a concise and efficient way to format strings. They allow direct variable embedding and expression evaluation inside curly braces `{}`.

````python
name = "Alice"
age = 25
````

````python

print(f"My name is {name} and I am {age} years old.")
````

````python
x = 5
y = 10
````

````python

print(f"The sum of {x} and {y} is {x + y}.")
````

#### Using the format() Method

The `format()` method (introduced in Python 2.7 and 3.0) allows inserting values into {} placeholders.

````python
name = "Bob"
age = 30
````

````python

print("My name is {} and I am {} years old.".format(name, age))
````

````python
print("{1} comes before {0}.".format("second", "first"))
````

Adding Elements to a List

Python provides three main ways to add elements to a list:

| Method    | Description                                      | Example                      |
|-----------|--------------------------------------------------|------------------------------|
| `append()` | Adds an element to the end of the list.         | `list.append(100)`          |
| `insert()` | Inserts an element at a specific position.      | `list.insert(1, "new")`     |
| `extend()` | Adds multiple elements from another iterable.   | `list.extend([200, 300])`   |


````python
fruits = ["apple", "banana"]
fruits.append("cherry")  # Adds "cherry" at the end
print(fruits)  # Output: ['apple', 'banana', 'cherry']

fruits.insert(1, "mango")  # Inserts "mango" at index 1
print(fruits)  # Output: ['apple', 'mango', 'banana', 'cherry']

fruits.extend(["grape", "orange"])  # Adds multiple elements
print(fruits)  # Output: ['apple', 'mango', 'banana', 'cherry', 'grape', 'orange']
````

Removing Elements from a List

Python provides three main ways to remove elements from a list:

| Method    | Description                                             | Example                      |
|-----------|---------------------------------------------------------|------------------------------|
| `remove()` | Removes the first occurrence of a specific value.      | `list.remove("apple")`      |
| `pop()`    | Removes and returns an element at a given index.       | `list.pop(2)`               |
| `del`      | Deletes an element at a specific index or the entire list. | `del list[1]` or `del list` |



````python
colors = ["red", "blue", "green", "yellow"]

colors.remove("blue")  # Removes "blue"
print(colors)  # Output: ['red', 'green', 'yellow']

colors.pop(1)  # Removes element at index 1 ("green")
print(colors)  # Output: ['red', 'yellow']

del colors[0]  # Deletes the first element
print(colors)  # Output: ['yellow']
````

Sorting and Reversing a List

Python provides built-in methods to sort and reverse lists.

| Method            | Description                                    | Example                      |
|-------------------|------------------------------------------------|------------------------------|
| `sort()`         | Sorts the list in ascending order (default).   | `list.sort()`               |
| `sort(reverse=True)` | Sorts the list in descending order.        | `list.sort(reverse=True)`   |
| `reverse()`      | Reverses the order of elements.                | `list.reverse()`            |

````python
numbers = [5, 2, 9, 1, 7]

numbers.sort()  # Sorts in ascending order
print(numbers)  # Output: [1, 2, 5, 7, 9]

numbers.sort(reverse=True)  # Sorts in descending order
print(numbers)  # Output: [9, 7, 5, 2, 1]

numbers.reverse()  # Reverses the list
print(numbers)  # Output: [1, 2, 5, 7, 9]
````

Set Operations

Python provides several operations to perform set manipulations, including union and intersection.

*Union (| or union())* -- The union of two sets returns a new set containing all unique elements from both sets.

````python
set1 = {1, 2, 3}
set2 = {3, 4, 5}


````

````python
union_set = set1 | set2  # OR set1.union(set2)
print(union_set)  # Output: {1, 2, 3, 4, 5}
````

*Intersection (& or intersection())* -- The intersection of two sets returns a new set containing only the elements that are common in both sets.

````python
set1 = {1, 2, 3}
set2 = {3, 4, 5}

````

````python

intersection_set = set1 & set2  # OR set1.intersection(set2)
print(intersection_set)  # Output: {3}
````

3 Dictionaries in Python

A **dictionary** is an unordered collection of **key-value pairs**. It is useful for **mapping and storing data efficiently**. There are several manipulations you can perform on a dictionary.



- Accessing Values in a Dictionary

We can retrieve a value by using its corresponding key inside **square brackets `[]`**.  
The `get()` method retrieves a value for a key and returns `None` if the key does not exist.

```python
# Creating a dictionary
student = {"name": "Alice", "age": 25, "course": "Data Science"}

# Accessing values
print(student["name"])  # Output: Alice
print(student.get("age"))  # Output: 25

# Using get() for a non-existent key
print(student.get("grade"))  # Output: None (prevents an error)


````python
# Creating a dictionary
student = {"name": "Alice", "age": 25, "course": "Data Science"}
````

````python

# Accessing values
print(student["name"])  # Output: Alice
print(student.get("age"))  # Output: 25

````

````python
# Using get() for a non-existent key
print(student.get("grade"))  # Output: None (prevents an error)
````

Adding and Updating Values

In Python, we can add new key-value pairs or update existing values using assignment `(=)` or the `update()` method.

````python
# Adding a new key-value pair
student["grade"] = "A"

# Updating an existing value
student["age"] = 26

# Using update() to modify multiple values
student.update({"name": "Bob", "course": "AI"})

print(student)
# Output: {'name': 'Bob', 'age': 26, 'course': 'AI', 'grade': 'A'}
````

Iterating Through a Dictionary

By default, looping through a dictionary iterates over the keys.

To access both keys and values simultaneously, use the `items()` method.
To loop through only the values, use the `values()` method.

````python
# Iterating through keys
for key in student:
    print(key)  # Output: name, age, course, grade

# Iterating through values
for value in student.values():
    print(value)  # Output: Bob, 26, AI, A

# Iterating through both keys and values
for key, value in student.items():
    print(f"{key}: {value}")
# Output:
# name: Bob
# age: 26
# course: AI
# grade: A
````

Removing Elements

`pop()`: Removes a specific key and returns its value.

`del`: Deletes a key-value pair permanently.

`popitem()`: Removes and returns the last added key-value pair as a tuple.

````python
# Removing an element using pop()
removed_value = student.pop("grade")
print(removed_value)
print(student)
````

````python

# Removing an element using del
del student["age"]
print(student)
````

````python
# Removing the last inserted key-value pair using popitem()
last_item = student.popitem()
print(last_item)
print(student)
````

4 Tuples in Python

A **tuple** is an **immutable**, ordered collection of elements.  
Once created, **you cannot modify** a tuple (no adding, removing, or changing elements).  
However, tuples support **indexing, slicing, and conversion operations**.

Accessing Elements in a Tuple

Elements in a tuple can be accessed using **indexing** and **slicing**.  
Each element in a tuple has an **index**, starting from `0` for the first element.

```python
# Creating a tuple
fruits = ("apple", "banana", "cherry", "date")

# Accessing elements using indexing
print(fruits[0])  # Output: apple
print(fruits[2])  # Output: cherry

# Accessing elements using slicing
print(fruits[1:3])  # Output: ('banana', 'cherry')
print(fruits[:2])  # Output: ('apple', 'banana')
print(fruits[-1])  # Output: date (last element)


````python
fruits = ("apple", "banana", "cherry", "date")

# Accessing elements using indexing
print(fruits[0])  # Output: apple
print(fruits[2])  # Output: cherry
````

````python

# Accessing elements using slicing
print(fruits[1:3])  # Output: ('banana', 'cherry')
print(fruits[:2])  # Output: ('apple', 'banana')
print(fruits[-1])  # Output: date (last element)
````

Tuple Unpacking

Tuple unpacking allows assigning elements of a tuple to multiple variables in a single step.

````python
# Tuple unpacking
person = ("Alice", 25, "Data Scientist")


````

````python
name, age, profession = person
print(name)  # Output: Alice
print(age)  # Output: 25
print(profession)  # Output: Data Scientist

````

````python
# Using underscore (_) as a placeholder for unwanted values
person_info = ("Bob", 30, "Engineer", "New York")

````

````python
name, age, _, city = person_info
print(name)  # Output: Bob
print(city)  # Output: New York
````

Converting Between Lists and Tuples

Python provides built-in functions to convert a list to a tuple and a tuple to a list.

This is useful when you need immutability (tuples) or mutability (lists) in different situations.

````python
# Converting a list to a tuple
list_numbers = [1, 2, 3, 4, 5]
tuple_numbers = tuple(list_numbers)

print(tuple_numbers)  # Output: (1, 2, 3, 4, 5)

````

````python

# Converting a tuple to a list
tuple_fruits = ("apple", "banana", "cherry")
list_fruits = list(tuple_fruits)

print(list_fruits)  # Output: ['apple', 'banana', 'cherry']
````

### Zip and Enumerate Iterators

1. `zip()`: Combining Multiple Iterables

The `zip()` function in Python allows you to combine multiple iterables (such as lists, tuples, or strings) into a single iterator.

syntax:

````python
zip(iterable1, iterable2, ...iterableN)
````

The returned result is a tuple containing elements from the corresponding positions of each iterable passed into the `zip()` function.

If the iterables have different lengths, `zip()` stops at the shortest iterable.

The result can also be converted into a list or dictionary if needed.

````python
names = ["Alice", "Bob", "Charlie"]
ages = [25, 30, 35]

````

````python
combined = zip(names, ages)
````

````python
combined
````

````python
list(combined)
````

````python
dict(combined)
````

````python
tuple(combined)
````

````python
set(combined)
````

````python
names = ["Alice", "Bob", "Charlie"]
ages = [25, 30]
````

***The third element in names is ignored because ages has fewer elements.***

#### Unpacking Zipped Iterables

You can "unzip" a zipped object using `zip(*zipped_object)`

````python
names = ["Alice", "Bob", "Charlie"]
ages = [25, 30]
````

````python

combined = zip(names, ages)
````

````python
names, ages = zip(*combined)
print(names)
print(ages)
````

````python
combined = zip(names, ages)
res = list(combined)
````

````python
res
````

````python
names, ages = zip(*res)
print(names)
print(ages)
````

### List Comprehension

List comprehension is a concise and efficient way to create lists in Python. It provides a more readable alternative to using loops for constructing lists.

syntax:

````python
[expression for item in iterable if condition]
````

- Expression: The operation performed on each item.

- Iterable: The data source (e.g., a list, range, or string).

- Condition (Optional): A filtering condition to select specific elements.

````python
for x in range(5):
  print(x)
````

````python
numbers = [x for x in range(5)]
````

````python
print(numbers)
````

- x for x in range(5): Iterates through values 0 to 4.

- Collects each value into the numbers list.

You can apply an operation to modify each element before adding it to the list:

````python
squared_numbers = [x**2 for x in range(5)]
````

````python
squared_numbers
````

You can also filter elements in List comprehension using an `if` condition:

image.png

````python
even_numbers = [x for x in range(10) if x % 2 == 0]
````

````python
even_numbers
````

````python
nested_list = [[1, 2], [3, 4], [5, 6]]
````

````python
flattened = [num for sublist in nested_list for num in sublist]
````

````python
print(flattened)
````

- Loop Control Flow Statements

Python provides additional statements to control loops:

`break`  Exits the loop completely.

`continue`  Skips the current iteration and moves to the next.

`pass`  Does nothing (a placeholder).

image.png

 [img source](https://stackoverflow.com/questions/9483979/is-there-a-difference-between-pass-and-continue-in-a-for-loop-in-python)

Using `break` to stop a loop early.

````python
for num in range(1, 10):
    if num == 5:
        break  # Stops the loop when num is 5
    print(num)
````

Using `continue` to skip an iteration.

````python
for num in range(1, 6):
    if num == 3:
        continue  # Skips number 3
    print(num)
````

Using `pass`
The `pass` statement does nothing. It is used as a placeholder when a block of code is required syntactically but you dont want to execute anything yet.



````python
for num in range(1, 6):
    if num == 3:
        pass  # Placeholder for future code
    else:
        print(num)
````
