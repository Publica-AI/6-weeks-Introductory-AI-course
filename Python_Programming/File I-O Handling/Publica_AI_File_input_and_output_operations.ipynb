###  Opening and Reading Files in Python

Python uses the `open()` function to access files. The basic syntax is:

```python
file = open("example.txt", "r")  # "r" means read mode
```

```python
content = file.read()
```

```python
print(content)
```

```python
file.close()  # Always close the file after reading
```

Using `with open()`, Python automatically closes the file after reading.

```python
with open("example.txt", "r") as file:

    content = file.read()

    print(content)  # No need to manually close the file
```

#### Reading Line by Line

```python
with open("example.txt", "r") as file:
    for line in file:

        print(line.strip())

```

### Writing to a File in Python

If you want to save data into a file, use `write (w)` or `append (a)` mode.

#### Write Mode `(w)` - Overwrites File

```python
with open("example.txt", "w") as file:

    file.write("Hello, Python!")
```

#### Append Mode `(a)` - Adds to the File

```python
with open("example.txt", "a") as file:

    file.write("\nWelcome to Python file handling!")
```

#### Writing Multiple Lines

```python
lines = ["Line 1\n", "Line 2\n", "Line 3\n"]

with open("example.txt", "w") as file:
    file.writelines(lines)  # Writes multiple lines at once
```

### Handling Different File Paths

When working with files, you need to specify their location (path). There are two types:

- Absolute Path  Full location of the file (e.g., "C:/Users/YourName/Documents/file.txt").
- Relative Path  Refers to the file based on the current directory (e.g., "file.txt").

image.png

#### Using os.path for File Paths

```python
import os

file_path = os.path.join("folder", "subfolder", "example.txt")
```

```python
print(file_path)
```

#### Using pathlib for Cleaner Paths



```python
from pathlib import Path

file_path = Path("folder") / "subfolder" / "example.txt"
```

```python
print(file_path)
```

### Creating and Managing Directories

If your program needs to save reports or logs, you may want to create folders dynamically.

#### Creating a Folder Using os

```python
import os

os.makedirs("new_folder", exist_ok=True)  # Creates the folder if it doesnt exist
```

#### Creating a Folder Using pathlib

```python
from pathlib import Path

Path("new_folder").mkdir(parents=True, exist_ok=True)
```

### Copying, Moving, and Deleting Files

Python provides shutil and os modules for file management.

#### Copying a File

```python
import shutil

shutil.copy("example.txt", "backup.txt")
```

#### Moving a File

```python
shutil.move("backup.txt", "archive/backup.txt")
```

#### Deleting a File

```python
import os

os.remove("example.txt")  # Deletes the file
```
