###### - Default Arguments
A function can have default values for parameters, so if no argument is provided, it uses the default.

example: When inviting people to a party, if you dont specify the dress code, its assumed to be "any party dress".

```python
def invite(name, dress_code="party dress"):
    print(f"Dear {name}, you are invited to the party. Dress code: {dress_code}.")
```

```python
invite("Ada")
```

```python
invite("Bola", "Traditional")
```

Here, "party dress" is the default argument for dress_code.



 ###### - Multiple Positional Arguments(*args)

`*args` is used when we want a function that can accept multiple inputs?

Think of `*args` like a bucket that collects all extra arguments into a tuple.

```python
def sum_numbers(*args):
    total = sum(args)
    print(f"Total sum: {total}")


```

```python
sum_numbers(2, 4, 6)
```

```python
sum_numbers(10, 20, 30, 40)
```

###### - Multiple Keyword Arguments( **kwargs )

While `*args` collects extra positional arguments as a tuple, `**kwargs` collects extra named arguments as a dictionary.

```python
def student_info(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")


```

```python
student_info(name="Emeka", age=21, course="Computer Science")
```

`**kwargs` allows us to pass multiple named arguments.

So far, weve seen functions that print values. But sometimes, we need a function to return a value instead of just displaying it. This is where the `return` statement comes in.

######  Multiple `return` Values

Python allows functions to return multiple values.

```python
def student_scores(name, math, english, science):
    total = math + english + science
    average = total / 3
    return name, total, average
```

### Local vs. Global Scope in Python

image.png

In Python, scope determines where a variable can be accessed.

###### Local Scope Variables

When we refer to local scope variables, we basically mean variables that can be accessed only within a block of code, in this case, function.
A local variable exists only inside a function.

```python
def show_balance():

    balance = 5000

    print(f"Your balance is {balance}")
```

```python
show_balance()
```

```python
print(balance)  #This will return an error because this variable, `balance`  is only accessible inside the function
```

###### Global Scope Variables

A global scope variable is defined outside a function and can be accessed anywhere.

```python
balance = 10000 # Global variable

def check_balance():
    print(f"Your balance is {balance}")


```

```python
check_balance()
```

```python
print(balance)
```

###### Modifying Global Variables Inside a Function

To modify a global variable inside a function, we use the `global` keyword.

```python
count = 0

def increase():
    global count  # This enables us to modify the global variable
    count += 2
    print(f"Count: {count}")

```

```python

increase()
```

```python
increase()
```

```python
print(count)
```

### Lambda Functions (Anonymous Functions)


A lambda function is a small, anonymous function in Python. Unlike normal functions defined with def, lambda functions are written in a single line and don't have a name.

Think about how Nigerians use shortened names:

Instead of saying "Oluwafemi", we say "Femi".
Instead of "Chukwuebuka", we say "Ebuka".
Both the full name and the short name refer to the same person, but just like `lambda` is a shorter way to define a function!

image.png

 Syntax of `Lambda` Functions

```python
lambda arguments: expression
```

- `lambda` keyword defines the function.
- It doesn't use return; the expression is automatically returned.
- Works best for short functions that fit in one line.

Example: To double a number using Python

- Using `def` function

```python
def double(x):
    return x * 2

```

```python

print(double(5))
```

- Using lambda

```python
double = lambda x: x * 2

```

```python
print(double(5))
```

 Both functions do the same thing, but lambda is shorter

Example: Adding Two Numbers

```python
add = lambda a, b: a + b

```

```python
print(add(4, 6))
```

When Should You Use lambda?
- When you only need a function temporarily and dont want to define it separately.
- When using functions like map(), filter(), and reduce() (explained next!).


When Shouldn't You Use lambda?
- Dont use lambda if the function is long or complexit reduces readability.

#### Using `map()`, `filter()`, and `reduce()`

1. `map()`  Applying a Function to Each Item

Imagine you're given a list of prices in dollars and you want to convert them to naira (). You apply the exchange rate to each itemthis is exactly what `map()` does!



```python
# assuming 1 dollar = 1500 naira
prices_in_dollars = [10, 20, 30]
exchange_rate = 1500


```

```python
prices_in_naira = list(map(lambda x: x * exchange_rate, prices_in_dollars))
```

```python
print(prices_in_naira)
```

-  Squaring Numbers

```python
numbers = [1, 2, 3, 4]
```

```python
squared = list(map(lambda x: x ** 2, numbers))
```

```python
print(squared)
```

2.  `filter()`  Selecting Items Based on a Condition

 `filter()` is like screening job applicants. If you're hiring only people with at least 3 years of experience, filter() will return only those who qualify.

```python
applicants = [("John", 5), ("Mary", 2), ("David", 3)]
```

```python
qualified = list(filter(lambda x: x[1] >= 3, applicants))
```

```python
print(qualified)
```

- Filtering Adults from a List of Ages

```python
ages = [15, 22, 17, 30, 45]

```

```python
adults = list(filter(lambda age: age >= 18, ages))
```

```python
print(adults)
```

3. `reduce()`  Cumulative Calculation

Imagine you are counting all the votes from different polling units to get the total result. reduce() is used to accumulate values like this case.

```python
from functools import reduce

votes = [120, 340, 200, 180]
```

```python
total_votes = reduce(lambda x, y: x + y, votes)
```

```python
print(total_votes)
```

- Summing a List of Numbers

```python

numbers = [1, 2, 3, 4]
```

```python
total = reduce(lambda x, y: x + y, numbers)
```

```python

print(total)
```

### Function Decorators

image.png

A decorator is a special function that modifies another function without changing its actual code.

Basic Syntax of a Decorator

```python
def decorator_name(func):
    def wrapper():
        # Do something before calling func()
        func()
        # Do something after calling func()
    return wrapper
```

- func() is the original function.
- The wrapper() function modifies func() before and after execution.

Example: Measuring Execution Time

A decorator can be used to measure how long a function takes to run.

```python
import time

def timer(func):
    def wrapper():
        start_time = time.time()  # Get start time
        func()  # Execute the function
        end_time = time.time()  # Get end time
        print(f"Execution time: {end_time - start_time:.5f} seconds")
    return wrapper
```

```python
@timer
def slow_function():
    time.sleep(2)  # Simulate a slow operation
    print("Function finished running.")

```

```python

slow_function()
```

 Example: Restricting Access

 Imagine a banking system where only logged-in users should access a function.

```python
def authenticate(func):
    def wrapper():
        user_logged_in = False  # Change to True to allow access
        if user_logged_in:
            func()
        else:
            print("Access Denied! Please log in.")
    return wrapper
```

```python

@authenticate
def check_balance():
    print("Your balance is 500,000.")
```

```python
check_balance()
```
