### Print Statement






What is the `print()` Statement?
The `print()` Statement in Python is used to display output. It helps us see the results of our code and debug issues.

Think of `print()` like a loudspeaker. it announces whatever you tell it to announce to the computer.

image.png








The syntax is simple:

You put whatever you want to display inside the parentheses. If it's text (a string), it should be inside quotes ("" or '').

```python
print("message")   #run this by pressing the play button by the left side of this code box
```

```python
print("Hello, world!")
```

You can print multiple values by separating them with commas (,)

```python
print("My name is", "Chisom")
print("I am learning", "Python", "for Data Science")
```

Python allows printing numbers directly without quotes.

```python
print(25)        # Prints a number
print(3.14)      # Prints a decimal (float)
print("Age:", 21) # Prints text and a number together
```

By default, print() separates multiple items with a space. But you can change this using sep.

```python
print("Python", "is", "awesome", sep="-")
```

### Adding Comments in Python

In the previous code block, we wrote:

```python
print(25)        # Prints a number
print(3.14)      # Prints a decimal (float)
print("Age:", 21) # Prints text and a number together
```

While learning how to print, we used the `#` symbol.

You must have wondered what this is. 🤔

Well, that `#` symbol is used to add comments in Python!

**What Are Comments?**
Comments are lines of text in your code that Python ignores when running the program. They are meant to provide more understanding, contexts and explanation to your code!

💡 Why use comments?
✔️ To explain your code for yourself and others.
✔️ To temporarily disable some lines of code.
✔️ To improve readability.



image.png

[img source](https://www.youtube.com/watch?v=90noStdV788)

- Single-Line Comments

In Python, a single-line comment starts with `#`. Here, Python ignores everything after `#`.

```python
# This is a comment
print("Hello, world!")  # This prints a message
```

- Multi-Line Comments

Python doesn’t have a built-in multi-line comment feature, but we can use multiple `#` symbols

```python
# This is a comment
# spread across
# multiple lines
print("Python is fun!")
```

- Alternative: Using Triple Quotes (""" """ or ''' ''')

Although triple quotes are mainly for docstrings (to be explained later), they can be used as multi-line comments:

```python
"""
This is a multi-line comment.
Python will ignore everything inside.
"""
print("Welcome to Python!")
```

- Commenting Out Code
You can disable a line of code using `#`.

```python
# print("This won't run")
print("This will run")
```

### Variables In Python

A variable is like a storage box where we keep data. It holds a value that can change or be reused throughout a program.

image.png

[img source](https://www.packetswitch.co.uk/python-variables-data-types/)

***Why do we need variables?***

Instead of writing the same value multiple times, we can store it in a variable and use it whenever needed.

For example, instead of writing:

```python
print("My name is John")
print("My name is not John")
```

We can use a variable:

```python
name = "John"
print("My name is", name)
print("My name is not", name)
```

This makes our code cleaner, reusable, and easier to modify!

**Declaring Variables in Python**

In Python, to declare a variable, just assign a value to the variable name!

```python
# Example: Creating variables

name = "Alice"
# Printing the variable
print(name)
```

```python
age = 25
# Printing the variable
print(age)
```

```python
height = 5.6
# Printing the variable
print(height)
```

```python
is_student = True
# Printing the variable
print(is_student)
```

**Variable Naming Conventions in Python**

Although Python doesn’t enforce strict naming rules, following conventions makes your code more readable.

Here are the common naming styles in Python:



| Naming Style            | Example    | Description                                         |
|-------------------------|-----------|-----------------------------------------------------|
| **Snake Case (Recommended)** | `user_name` | Uses lowercase letters with underscores (_) between words. |
| **Camel Case**          | `userName` | First word is lowercase, next words start with uppercase. |


```python
# Using snake case (Recommended for variables)
first_name = "Alice"
last_name = "Johnson"
```

```python
# Using camel case
piNumber = 3.14159
maxAttempts = 5
```

🚀 Best Practice:

✅ Use descriptive variable names (e.g., user_age instead of x).

✅ Stick to snake_case for regular variables.

**Variable Naming Rules**

While naming a variable, follow these rules:

✅ Can start with a letter or underscore _

✅ Can contain letters, numbers, and underscores

✅ Are case-sensitive (Age and age are different)

❌ Cannot start with a number

❌ Cannot use Python keywords (like print, if, while, etc.)


```python
# Examples of valid variable names:
first_name = "John"
age = 30
height_cm = 175
```

```python
# Examples of invalid variable names:

2name = "Alice"    # ❌ Starts with a number
first name = "Ben" # ❌ Contains space
if = "hello"       # ❌ Uses a reserved keyword
```

-  3️⃣ Comparison Operators (For Comparing Values)
These operators compare two values and return True or False.

Comparison Operators in Python includes:

1. `==` → Equal to
2. `!=` → Not equal to
3. `>` → Greater than
4. `<` → Less than
5. `>=` → Greater than or equal to (True if at least one condition is satisfied.)
6. `<=` → Less than or equal to (True if at least one condition is satisfied.)

```python
a = 7
b = 10

print(a == b)  # False (7 is not equal to 10)
```

```python
print(a != b)  # True  (7 is not equal to 10)
```

```python
print(a > b)   # False (7 is not greater than 10)
```

```python
print(a < b)   # True  (7 is less than 10)
```

```python
print(a >= 7)  # True  (7 is equal to 7)
```

```python
print(b <= 10) # True  (10 is equal to 10)
```

- 4️⃣ Logical Operators (For Decision-Making)
These operators are used to combine multiple conditions.

Logical Operators in Python includes:

1. and → Returns True if both conditions are True.
2. or → Returns True if at least one condition is True.
3. not → Reverses the condition (True becomes False and vice versa).

```python
# Example:

x = 5
y = 10

print(x > 2 and y < 15)  # True (Both are true)
```

```python
print(x > 2 or y > 15)   # True (One condition is true)
```

```python
print(not (x > 2))       # False (Negates the result)
```

### Basic Data Types & Structures in Python

In Python, data comes in different forms. Some values are numbers, some are text, and others may be collections of multiple values.

Python provides data types to define the kind of values we work with and data structures to organize multiple values efficiently.

#### **Data Types**



Python has several built-in data types:

| **Data Type**            | **Example**   | **Description**  |
|--------------------------|--------------|------------------|
| **Integer (`int`)**      | `10`         | Represents whole numbers (integers). |
| **Float (`float`)**      | `3.14`       | Represents decimal numbers (floating-point numbers). |
| **String (`str`)**       | `"Hello"`    | Represents text (a sequence of characters). |
| **Boolean (`bool`)**     | `True` / `False` | Represents Boolean values (True or False). |


```python
# Example:

# Integer
age = 25
print(age)  # Output: 25
```

```python
type(age)
```

```python
# Float
pi = 3.14
print(pi)  # Output: 3.14
```

```python
type(pi)
```

```python
# String
name = "Alice"
print(name)  # Output: Alice
```

```python
type(name)
```

```python
# Boolean
is_python_fun = True
print(is_python_fun)  # Output: True
```

```python
type(is_python_fun)
```

#### **Data Structures**



Python also has data structures that allow us to store and organize multiple values efficiently.

| **Data Structure**       | **Example**                    | **Description**  |
|--------------------------|--------------------------------|------------------|
| **List (`list`)**        | `[1, 2, 3]`                    | Ordered, changeable collection of items. |
| **Tuple (`tuple`)**      | `(4, 5, 6)`                    | Ordered, immutable collection of items. |
| **Set (`set`)**          | `{7, 8, 9}`                    | Unordered collection of unique items. |
| **Dictionary (`dict`)**  | `{"name": "Bob", "age": 30}`   | Collection of key-value pairs. |


```python
# Example:
```

```python
# List (ordered and changeable)
fruits = ["apple", "banana", "cherry"]
print(fruits)
```

```python
type(fruits)
```

```python
# Tuple (ordered but immutable)
coordinates = (10, 20)
print(coordinates)
```

```python
type(coordinates)
```

```python
# Set (unordered and unique)
unique_numbers = {1, 2, 3, 2, 1}
print(unique_numbers)
```

```python
type(unique_numbers)
```

```python
# Dictionary (key-value pairs)
person = {"name": "Bob", "age": 30}
print(person)
```

```python
type(person)
```
