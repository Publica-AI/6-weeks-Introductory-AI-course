**Types Of Visuals  And Their Usecases**

1. **Line Plot**: This is used to show trends or changes over time. You can use it if you want to track changes.

2. **Bar Plot**: This is used to compare quantities between categories.

3. **Scatter Plot**: To show relationship between two variables.
4. **Histogram**: To show the distribution of your data.

5. **Piechart**: To show proportions or percentages of a whole.
6. **Boxplot**: Primarily for detecting outliers, show distribution and the spread of your data.

7. **Heatmap**: This is used for correlation, showing the relationships between your data variables through colour intensity.

Lets get started...

* To use matplotlib library, you have to import it into your environment.

```
import matplotlib.pyplot as plt
```
#### Functions in Matplotlib

* Creating your plots and charts
```
plt.plot() # This is for line chart
plt.bar() # This is used for bar chart
plt.scatter() # This is used for scattered plots
plt.hist() # This is used for plotting histogram
plt.boxplot() # This is used for creating boxplots
plt.pie() # This is creating pie charts
plt.imshow() # This is used for creating heatmap(correlation)
```

* customizing and setting your axis and labels
```
plt.title() # the title of your plot
plt.xlabel() # the x axis label
plt.ylabel() # the y axis label
plt.xlim() # limit for your x axis
plt.ylim() # limit for your y axis
plt.xticks() # the orientation of your x- values
plt.yticks() # the orientation of your y- values
plt.show()
```

* Saving your plots

```
plt.show()
plt.savefig()
```

* Adding legends to your plot

```
plt.legend()
plt.grid()
```
* Add figures and Subplots functions

```
plt.figure()
plt.subplot()
plt.subplots() # for multiple subplots
plt.tight_layouts()
```
