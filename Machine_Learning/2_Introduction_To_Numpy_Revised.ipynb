**NumPy (Numerical Python)** is a powerful library for numerical computing in Python. It provides support for large, multi-dimensional arrays and matrices, along with mathematical functions to operate on these arrays efficiently.

Python lists are flexible but slow for numerical computations. NumPy arrays are:Faster,Memory Efficient, Convenient, Supports Vectorization etc.

```python
# To use numpy the first step is to import it import numpy as alias
%%time
import numpy as np
import time

# Experiment to see the speed of numpy
size = 1000000


py_list = list(range(size))
#start = time.time()
py_result = [x * 2 for x in py_list]
#end = time.time()
#print("Python list time:", end - start)
```

```python
%%time
np_array = np.arange(size)
np_result = np_array * 2
#print("NumPy array time:", end - start)
```

Look at the time and note the difference... You will notice how fast numpy is.

### **Creating Numpy Arrays**

1D Array (Vector)

```python
# Let create numpy arrays

array1 = np.array([1, 2, 3])
print(array1)
# From the output, we could see that this is a 1D array

print(array1.shape) # output showed 3 column, no row.
```

2D Array (Matrix)

```python
# Lets create a 2D array
array2 = np.array([[1, 2, 3], [4, 5, 6]])

print(array2)
print(array2.shape) # output showed 3 columns, 2 rows
```

3D Arrays (Tensor)
* Refer back on the 3D image in the slide on numpy array.

```python
# Lets create a 3D array

array3 = np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])
print(array3)
print(array3.shape) # output showed 2 columns, 2 rows, 2 depth
```

```python
# Checking the properties of Arrays

print("Number of Dimensions:", array3.ndim)  # 1D, 2D, or 3D?
print("Total Elements:", array3.size)  # Total number of elements
print("Data Type:", array3.dtype)  # Data type of elements
```

Indexing

```python
# Lets define an array
access_array = np.array([10, 20, 30, 40, 50])

# LEts get the first element
print(access_array[0])

# LEts get  the last element (you recall negative indexing?)
print(access_array[-1])

# Ensure to observe the output as you code along...
```

### Modifying Elements

One thing that makes numpy unique, is that it supports vectorized operations

That is, the operationscan be applied to all elements of an array simultaneously without the need for loops.

```python
# LEts define an array
vectorised_array = np.array([100, 200, 300, 400, 500, 600, 700, 800])

# LEts add 50 to each element and take a look at the output
print(vectorised_array + 50)
```

```python
# We are adding this, juast like we add vectors in mathematics...Did you still remember that topic?

arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])

# Lets add
print(arr1 + arr2)

# Lets multiply
print(arr1 * arr2)

# Lets divide
print(arr1 / arr2)

# Oberve the output and replicate the same thing...
# You can decide to be creative, you know..
```

### Reshapin g Arrays

You can reshape arrays to change their dimensions (e.g., converting a 1D array to a 2D array)
When we get to deep learning we would see the importance of using numpy arrays to reshape 3D images.

For now lets keep it simple and grasp the concept first.

```python
arr = np.array([1, 2, 3, 4, 5, 6])

# lets reshape to 2 by 3 array (2 rows, 3 columns)
reshaped_arr = arr.reshape(2, 3)
print(reshaped_arr)
```

```python
# We can also flatten multi-dimensional to 1 dimension.
flattened_arr = reshaped_arr.flatten()

print(flattened_arr)
print(flattened_arr.shape)
```

```python
# Solutions

print("\nSolution 1")
# 1. Create a 6x6 array with random integers between 1 and 50
arr = np.random.randint(1, 50, (6, 6))
print("Original Array:\n", arr)

print("\nSolution 2")
# 2. Extract required parts
print("\nFirst two rows:\n", arr[:2])
print("\nLast three columns:\n", arr[:, -3:])
print("\nAll even numbers:\n", arr[arr % 2 == 0])

print("\nSolution 3")
# 3. Reshape and Flatten
reshaped_arr = arr.reshape(3, 2, 6)
print("\nReshaped to (3,2,6):\n", reshaped_arr)
flattened_arr = reshaped_arr.flatten()
print("\nFlattened back to 1D:\n", flattened_arr)

print("\nSolution 4")
# 4. Identity matrix with custom diagonal
identity_matrix = np.eye(4) * np.arange(10, 50, 10)
print("\nIdentity Matrix with Custom Diagonal:\n", identity_matrix)

print("\nSolution 5")
# 5. Matrix multiplication
mat1 = np.random.randint(1, 10, (3, 4))
mat2 = np.random.randint(1, 10, (4, 3))
product = np.dot(mat1, mat2)
print("\nMatrix 1:\n", mat1)
print("\nMatrix 2:\n", mat2)
print("\nMatrix Multiplication Result:\n", product)

print("\nSolution 6")
# 6. Function for transformation
def transform_array(arr):
    transformed = np.where(arr % 2 == 0, arr ** 2, -1)
    return transformed

test_arr = np.random.randint(1, 20, (4, 4))
print("\nOriginal Test Array:\n", test_arr)
print("\nTransformed Array:\n", transform_array(test_arr))
```
