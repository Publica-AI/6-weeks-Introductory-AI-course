# Descriptive Analysis

Overview

Descriptive statistics are used to summarize and analyze data to uncover patterns, distributions, and key metrics. It provides a foundation for understanding datasets before performing advanced statistical analysis or building predictive models.

To start, we import the necessary libraries for data manipulation, visualization, and statistical analysis:
- `pandas`: For data analysis and manipulation.
- `numpy`: For numerical computations.
- `matplotlib.pyplot` and `seaborn`: For creating visualizations.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

** Import dataset**

 We use the Titanic dataset, which is readily available in the seaborn library.

```python
# Import Titanic dataset (inbuilt in Seaborn library)
df = sns.load_dataset('titanic')
```

**Exploratory data analysis**

We will explore the data to gain insights about the data.

View dimensions of dataset


```python
print("Dataset Dimensions:", df.shape)
```

```python
# Preview first 5 rows of the dataset
print("\nDataset Preview:")
print(df.head())
```

- Use `df.head()` to view the first 5 rows.
- Use `df.info()` to examine data types and non-null values.

```python
# View summary of dataset
print("\nDataset Information:")
print(df.info())
```

```python
# Check for missing values
print("\nMissing Values:")
print(df.isnull().sum())
```

Using `df.describe()`, we can see the count, mean, standard deviation, min, and max values for numerical columns.

```python
# Descriptive statistics for numerical columns
print("\nSummary Statistics for Numerical Columns:")
print(df.describe())
```

Using `df.describe(include=['object'])`, we can summarize categorical columns to see:
- Unique categories.
- Most frequent category (`top`).
- Frequency of the most frequent category (`freq`).

```python
# Descriptive statistics for categorical columns
print("\nSummary Statistics for Categorical Columns:")
print(df.describe(include=['object']))
```

```python
# Descriptive statistics for all columns
print("\nSummary Statistics for All Columns:")
print(df.describe(include='all'))
```

Measures of Central Tendency
We compute:
- **Mean**: Average value using `df['age'].mean()`.
- **Median**: Middle value using `df['age'].median()`.
- **Mode**: Most frequent value using `df['age'].mode()`.
Here, we use the `age` column for demonstration.

```python
# Measures of central tendency for 'fare'
mean = df['age'].mean()
median = df['age'].median()
mode = df['age'].mode()

```

```python
print(f"Mean Age: {mean}")
print(f"Median Age: {median}")
print(f"Mode Age: {mode[0]}")
```

Lets visualize the Distribution of the Age Column
Using `sns.histplot to understand its spread and shape. let us also compute skewness and kurtosis for interpretation.

```python
sns.histplot(df['age'].dropna(), bins=10, kde=True)
plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

- Shape: The distribution is slightly right-skewed, meaning there's a tail extending towards older ages. It's not perfectly symmetrical (normal).

- Frequency: The tallest bars are in the 20-30 age range, indicating this is the most common age group.KDE Curve: The line (Kernel Density Estimate) provides a smoothed representation of the distribution, highlighting the skew and central tendency.  
- The age distribution is concentrated in the young adult range, with a tendency towards slightly older ages.

```python
# Measures of distribution shape
print(f"Skewness: {df['age'].skew()}")
print(f"Kurtosis: {df['age'].kurt()}")
```

Skewness: 0.389 This positive value confirms the visual right skew. It indicates that the tail on the right side of the distribution is longer than the tail on the left side.

Kurtosis: 0.178 This positive value indicates a slightly peaked distribution with heavier tails compared to a normal distribution. However, the kurtosis is relatively low, suggesting that the distribution is not drastically different from a normal distribution.

The positive skew suggests a moderate number of older individuals were present. The kurtosis indicates a slightly peaked distribution with relatively light tails, meaning there are fewer extreme outliers than in a normal distribution.

```python
sns.histplot(df['fare'], bins=20, kde=True, label='Fare', color='blue')
plt.title("Distribution of Fare")
plt.xlabel("Fare")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

Identifying Outliers with a Boxplot
A boxplot visualizes the spread of data and identifies potential outliers. Here, we use the `fare` column.

Most Titanic passengers paid lower fares, as shown by the tall bars on the left.  A few passengers paid much higher prices, represented by the long tail extending to the right. This reflects the different travel classes on the ship. The KDE curve visually emphasizes the skewness of the distribution, highlighting the concentration of fares at the low end.

```python
# Boxplot for Fare
plt.boxplot(df['fare'].dropna())
plt.title("Boxplot of Fare")
plt.ylabel("Fare")
plt.show()
```

Majority of Fares: The box is concentrated at the lower end of the fare range, indicating that most fares were relatively low.
Median Fare: The line inside the box suggests the median fare is also on the lower side.
Wide Range: The whiskers show a very wide range of fares, from very low to moderately high.
Numerous Outliers: The numerous circles above the upper whisker indicate a substantial number of unusually high fares.

```python
# Measures of dispersion for 'fare'
min_fare = df['fare'].min()
max_fare = df['fare'].max()
range_fare = max_fare - min_fare
var_fare = df['fare'].var()
std_fare = df['fare'].std()
q1 = df['fare'].quantile(0.25)
q2 = df['fare'].quantile(0.5)
q3 = df['fare'].quantile(0.75)
iqr = q3 - q1
```

```python
print(f"\nMinimum Fare: {min_fare}")
print(f"Maximum Fare: {max_fare}")
print(f"Range of Fare: {range_fare}")
print(f"Variance of Fare: {var_fare}")
print(f"Standard Deviation of Fare: {std_fare}")
print(f"Q1 (25th Percentile): {q1}")
print(f"Q2 (Median or 50th Percentile): {q2}")
print(f"Q3 (75th Percentile): {q3}")
print(f"IQR (Interquartile Range): {iqr}")
```

- Range: Fares varied wildly, from free (perhaps for crew or very young children) all the way up to about 512.
- Average Spread: The standard deviation (around 50) tells us how spread out the fares typically were from the average. A larger value means fares were more diverse.
- Typical Fares: The interquartile range (IQR, about 23) shows the range of the middle 50% of fares. In other words, half of all passengers paid somewhere between 8 and 31 for their tickets.

```python
# Measures of shape of distribution
skewness = df['fare'].skew()
kurtosis = df['fare'].kurt()
```

```python
print(f"\nSkewness of Fare: {skewness}")
print(f"Kurtosis of Fare: {kurtosis}")
```

- Skewness: 4.79 - This extremely high positive skewness indicates that the distribution of fares is heavily concentrated towards the lower end. In simpler terms, a large majority of passengers paid relatively low fares, while only a small fraction paid very high fares. This aligns with the class divisions on the Titanic, where 3rd class passengers paid significantly less than 1st class.

- Kurtosis: 33.40 - This exceptionally high kurtosis value suggests a distribution with very heavy tails and a sharp peak.  It means there's a high probability of observing extreme values (very high fares in this case) compared to a normal distribution. This reinforces the idea of vast differences in ticket prices between the classes, with a few extremely luxurious (and expensive) tickets skewing the distribution.
