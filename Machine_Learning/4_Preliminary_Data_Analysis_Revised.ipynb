### **Preliminary Data Analysis (PDA)**

PDA is the first, crucial step in the data analysis process. It involves exploring your data to understand its basic characteristics, detect potential issues, and guide further, more detailed analyses.

When you load in your data into the IDE, the first task to perform should be PDA so you can have a very good understanding of your data before continuing with your analysis.
